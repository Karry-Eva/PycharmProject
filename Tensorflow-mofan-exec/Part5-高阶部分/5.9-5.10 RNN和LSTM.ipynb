## 5.9 什么是循环神经网络RNN

【RNN流程】：每次 RNN 运算完之后都会产生一个对于当前状态的描述 , state. 我们用简写 S( t) 代替, 然后这个 RNN开始分析 x(t+1) , 他会根据 x(t+1)产生s(t+1), 不过此时 y(t+1) 是由 s(t) 和 s(t+1) 共同创造的.

## 5.10 什么是LSTM循环神经网络

优点：能解决RNN梯度消失和梯度爆炸的问题
特点：LSTM 和普通 RNN 相比, 多出了三个控制器. (输入控制, 输出控制, 忘记控制). 
- 输入gate：判断是否将X(t)进行写入（write）
- 忘记gate：要不要丢弃之前内存（memory）
- 输出gate：要不要输出这门（gate）
